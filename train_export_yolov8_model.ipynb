```python
from ultralytics import YOLO

# Cargar tu modelo custom
model = YOLO("/content/JUEGO.pt")

# Ver clases
labels = model.names  # esto es un diccionario {0: "clase0", 1: "clase1", ...}

# Convertir a lista ordenada
labels_list = [labels[i] for i in range(len(labels))]

print("Clases del modelo:")
for i, label in enumerate(labels_list):
    print(i, label)
```
